# Feature Engineering

```python
#Drop Name
data = data.drop("Name", axis=1)
data = data.drop("Name.1", axis=1)


#D.O.B. to Age (easier to use)
data["DOB"] = 2023 - data["DOB"]
data["DOB.1"] = 2023 - data["DOB.1"]
data.rename(columns = {'DOB':'Age', 'DOB.1':'Age.1'}, inplace = True)


cols = data.columns

for name in cols:
    #CHANGE NAMES TO _RED/BLUE
    if name[-2:] == ".1":
        new_name = name[0:-2]+"_RED"
    else:
        new_name = name+"_BLUE"
    data.rename(columns = {name:new_name}, inplace = True)
    name = new_name
    
    #REPLACE CATEGORIES W DUMMY VARIABLES
    if data[name].dtype =='object':
        index = data.columns.get_loc(name) #Cant enumerate since we're adding columns
        dummies = pd.get_dummies(data[name], prefix=name)
        left = data.iloc[:,0:index]
        right = data.iloc[:,index+1:]
        data = pd.concat([left,dummies,right],axis=1) #concat data back

        
data.head()
```

# Split Data
- Train: Create model parameters
- Valid: Estimate model performance + hyperparameter tuning
- Test: unseen data to do final bet analysis

```python
from sklearn.model_selection import train_test_split

#SPLIT: DATA --> TRAIN + VALID
index_train, index_valid  = train_test_split(np.array(data.index), train_size=0.7, random_state=1)
train = data.loc[index_train,:]
valid =  data.loc[index_valid,:]

#SPLIT: VALID --> TEST + VALID
index_valid, index_test  = train_test_split(np.array(valid.index), train_size=0.7, random_state=1)
test = valid.loc[index_test,:]
valid = valid.loc[index_valid,:]


#Split x and y variables
#y
y_train= train['Win_BLUE'].to_numpy()
y_valid = valid['Win_BLUE'].to_numpy()
y_test = test["Win_BLUE"].to_numpy()

#x
train = train.iloc[:,0:-1]
valid = valid.iloc[:,0:-1]
test = test.iloc[:,0:-1]

#Double check things went okay
total = len(data)
print("Total:",total)
print("TRAIN: Length ",len(train),": Percentage",round(len(train)/total,2))
print("VALID: Length ",len(valid),": Percentage",round(len(valid)/total,2))
print("TEST: Length",len(test),": Percentage",round(len(test)/total,2))
#print(len(train) + len(valid) + len(test))
```

# Function: Estimate Performance

```python
from sklearn.metrics import accuracy_score


def estimate_performance(models, name, train=train, valid=valid, y_train=y_train, y_valid=y_valid):
    tests = ["Accuracy"]
    results = pd.DataFrame(0.0, columns=tests, index=name)
    
    for number, model in enumerate(models):
        model.fit(train, y_train)
        predictions = model.predict(valid)
        
        accuracy = round(accuracy_score(y_valid, predictions),2)
        results.iloc[number, 0] = accuracy

    return results
```

# Untuned Results

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB


Logistic = LogisticRegression()
Boost = GradientBoostingClassifier()
KNN = KNeighborsClassifier()
SVC = SVC()
Tree = DecisionTreeClassifier()
Forest = RandomForestClassifier()
AdaBoost = AdaBoostClassifier()
NB = GaussianNB()

models = [Logistic, Boost, KNN, SVC, Tree, Forest, AdaBoost,NB]
name = ["Logistic Reg", "Gradient Boost", "KNN", "SVC", "Tree", "Forest", "Ada","Naive Bayes"]
results = estimate_performance(models, name)
results
```

# Tune Gradient Boost
- Optuna will test many iterations of hyperparameters and give back the best it found. Unlike a random search it will use previous results to inform future searches.

```python
run=False #If you set to false it'll just run a previously found optimum
```

```python
import optuna
from optuna.samplers import TPESampler
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

if run == True:
    def objective(trial):

        max_depth = trial.suggest_int('max_depth', 1, 20)
        n_estimators = trial.suggest_int('n_estimators', 1, 2000)
        #subsample = trial.suggest_int('subsample', 0,1)

    
        model = GradientBoostingClassifier(n_estimators = n_estimators,
                                           max_depth = max_depth,
                                           subsample = 0.5,
                                           random_state = 1)

        scores = cross_val_score(model, train, y_train, cv = 3,  scoring = 'accuracy')

        return np.mean(scores)

    sampler = TPESampler(seed=42) 
    study = optuna.create_study(direction='maximize', sampler=sampler)
    study.optimize(objective, n_trials = 1000, timeout = 1000) 
```

```python
if run == True:
    Grad_params = study.best_params
    print(Grad_params)
else:
    Grad_params = {'max_depth': 7, 'n_estimators': 1050}


Boosted =  GradientBoostingClassifier(learning_rate = 0.01, subsample = 0.5, **Grad_params, random_state=1)

models = [Boosted]
name = ["Gradient Boost Classifier"]
Boosted_Results = estimate_performance(models, name)
Boosted_Results
```

# Final Betting Model:
- Model has abillity to place a bet on red or blue. Also has abillity to not bet if too risky.
- 0 = bet on Red
- 1 = bet on Blue
- -1 = don't bet

```python
model = Boosted
risk_tolerance = 0.05 #between 0 and 0.5. Will not bet if prediction probabillity is between 1-risk_tolerance and 0+risk_tolerance
```

```python
predictions = model.predict_proba(test)
predictions = predictions[:,1]
bets = [round(prob) if prob < risk_tolerance or prob>(1-risk_tolerance) else -1 for prob in predictions]
```

```python
bets_taken = 0
bets_accurate = 0

for i in range(len(bets)):
    if bets[i] != -1:
        bets_taken +=1
        if bets[i] == y_test[i]:
            bets_accurate+=1
            
print("BETS TAKEN:")
print(len(bets),"Potential Bets")
print(bets_taken,"Bets Taken")
print(round(bets_taken/len(bets),2), "Percent Taken")
print("")

print("ACCURACY")
print(bets_accurate, "Accurate Bets")
print(round(bets_accurate/bets_taken,2),"Accuracy on Bets Taken")
print(round(accuracy_score(y_test, model.predict(test)),2),"Baseline Accuracy (on all fights)")
```
